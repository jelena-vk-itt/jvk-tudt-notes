# 10

```python
import scipy.stats as ss
print(ss.pearsonr(auto_weight, auto_df['mpg']))
```

# 11

```python
auto_ctab = pd.crosstab(auto_df['year'], auto_df['origin'])
auto_ctab
```

# 12

```python
auto_ctab7980 = auto_ctab.loc[[79, 80]]
auto_ctab7980
```

```python
chisq, p, dof, evs = ss.chi2_contingency(auto_ctab7980)
print("chi-squared " + str(chisq))
print("p = " + str(p))
print("degrees of freedom = " + str(dof))
print("expected value table: ")
print(evs)
```

The p-value is extremely low, indicating that there is probably a dependence between year and origin of car, i.e. that the popularity of cars from different origins changed between 1979 and 1980.

# 13

```python
chisq, p, dof, evs = ss.chi2_contingency(auto_ctab.loc[[80, 81]])
print("chi-squared " + str(chisq))
print("p = " + str(p))
print("degrees of freedom = " + str(dof))
print("expected value table: ")
print(evs)
```

The p-value would not indicate a correlation at the usual levels of 5% or 1%. It would indicate a correlation at a 10% level. We cannot be very confident that there is a correlation. In comparison with the previous result, this may not be surpirsing since the oil crisis in 1979 may have prompted people to look for more efficient cars, buying them from Europe and Japan.
